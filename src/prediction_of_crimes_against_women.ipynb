Here, we are using label encoder for encoding of categorical features instead of one hot encoder because from above command we can see that there are 36 unique values in STATE/UT column and 954 Unique values in DISTRICT column so, if we would have used one hot encoder than it would have generated a lot of unnecessary features which would later on affect our model and will cause heavy memory consumption.
```python
def with_hue(data,feature,ax):

    #Numnber of categories
    num_of_cat=len([x for x in data[feature].unique() if x==x])

    bars=ax.patches

    for ind in range(num_of_cat):
        ##     Get every hue bar
        ##     ex. 8 X categories, 4 hues =>
        ##    [0, 8, 16, 24] are hue bars for 1st X category
        hueBars=bars[ind:][::num_of_cat]
        # Get the total height (for percentages)
        total=sum([x.get_height() for x in hueBars])
        #Printing percentages on bar
        for bar in hueBars:
            percentage='{:.1f}%'.format(100 * bar.get_height()/total)
            ax.text(bar.get_x()+bar.get_width()/2.0,
                   bar.get_height(),
                   percentage,
                    ha="center",va="bottom",fontweight='bold',fontsize=10)

def without_hue(data,feature,ax):

    total=float(len(data))
    bars_plot=ax.patches

    for bars in bars_plot:
        percentage = '{:.1f}%'.format(100 * bars.get_height()/total)
        x = bars.get_x() + bars.get_width()/2.0
        y = bars.get_height()
        ax.text(x, y+2,y,ha='center',fontweight='bold',fontsize=7)
```

```python
def plotting_cat_features(nrows,ncols,cat_columns):

    f,ax=plt.subplots(nrows=nrows,ncols=ncols,figsize=(20,25))
    f.patch.set_facecolor('#ffffff')
    #Setting background and foreground color
    for i in range(0,nrows):
        for j in range(0,ncols):
            ax[i][j].set_facecolor('#ffffff')

    #Plotting count plot
    for i in range(0,nrows):
        for j in range(0,ncols):
            a1=sns.barplot(data=df1,x=df1.index,y=cat_columns[i*(nrows-2)+j],palette='viridis',ax=ax[i][j])
            without_hue(df,cat_columns[i-1],a1)
            #Dealing with spines
            ax[i][j].spines['top'].set_visible(False)
            ax[i][j].spines['right'].set_visible(False)
            ax[i][j].spines['left'].set_visible(False)
            ax[i][j].grid(linestyle="--",axis='y',color='gray')
```

## **5.2 Regression Model**

Here we would be trying to predict the number of recorded cases of any one crime activity say number of recorded rape cases with the help of other present features

```python
model_df.head()
```

Preparing the features and labels to fed to regression model and further splitting the dataset into training and testing set.

```python
features=model_df.drop(columns=["Rape"],axis=1)
labels=model_df.pop("Rape")
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,labels, test_size = 0.3, random_state = 42)
```

```python
names, mses, rmses, r2s, accuracy = [], [], [], [], []
```

### **Linear Regression**

Linear regression is a very basic model used for modelling the relationship between scalar response and one or more explanatory variables.

```python
#linear regression
lr = LinearRegression()
lr.fit(x_train, y_train)

# Prediction
pred = lr.predict(x_test)

# Performance Measure
lr_mse = mean_squared_error(y_test, pred)
lr_rmse = np.sqrt(lr_mse)
lr_r2 = r2_score(y_test, pred)

# Show Measures
result = '''
MSE  : {}
RMSE : {}
R^2  : {}
'''.format(lr_mse, lr_rmse, lr_r2)

print(result)

# Model Comparision
names.append("Linear Regression")
mses.append(lr_mse)
rmses.append(lr_rmse)
r2s.append(lr_r2)
```

Printing accuracy of model on training set and testing set

```python
lr_ac_train = lr.score(x_train,y_train)
lr_ac_test = lr.score(x_test,y_test)
accuracy.append(lr_ac_train)
(lr_ac_train,lr_ac_test)
```
